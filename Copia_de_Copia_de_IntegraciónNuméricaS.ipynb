# **Fórmulas de integración de Newton-Cotes**
Las *fórmulas* de *Newton-Cotes* son los tipos de integración numérica más comunes. Se basan en la estrategia de reemplazar una función complicada o datos tabulados por un polinomio de aproximación que es fácil de integrar:

$$ I = \int_a^b f(x) dx \cong \int_a^b f_n(x) dx$$

donde $f_n(x)$  es un polinomio de la forma

$$f_n(x) = a_0 + a_1x + ... + a_{n-1}x^{n-1} + a_nx^n$$

donde $n$ es el grado del polinomio (Chapra, S. C. y P. Canale, R. 2007. pp. 619).

La regla del rectángulo es una aproximación del área bajo la curva de una función en un intervalo $[𝑎, 𝑏]$, mediante la suma de $𝑛$
rectángulos, donde su base es definida por 
$Δ𝑥 = \dfrac{b-a}{n}$ y su altura está
dada por la evaluación de la función en los punto $𝑥_i=a+\Delta x*i$ de la partición del intervalo.
$$ A \approx \sum_{i = 1}^n f(x_i) \Delta x$$

##**Regla del Trapecio**

La regla del trapecio es la primera de las fórmulas cerradas de integración de Newton-
Cotes. Corresponde al caso donde el polinomio de aproximacion es de primer grado:

$$ I = \int_a^b f(x) dx \cong \int_a^b f_1(x) dx$$

donde

$$f_1(x) = f(a) + \frac{f(b) - f(a)}{b - a}(x - a)$$

El área bajo esta línea recta es una aproximación de la integral de $f(x)$ entre los límites $a$ y $b$:

$$f_1(x) = \int_a^b \left[f(a) + \frac{f(b) - f(a)}{b - a}(x - a)\right] dx$$

El resultado de la integración es

$$I = (b-a)\frac{f(a) + f(b)}{2} $$

que se denomina *regla del trapecio* (Chapra, S. C. y P. Canale, R. 2007. pp. 621).

extendiendo este procedimiento para el caso de aplicaciones multiples, se obtiene

$$ I = (b - a) \left(f(x_0) + 2 \sum_{i=1}^{n-1}f(x_i) + f(x_n)\right)\frac{1}{2n}$$

##**Regla de Simpson 1/3**

La regla de Simpson 1/3 resulta cuando un polinomio de interpolación de segundo grado
se sustituye en la ecuación

$$ I = \int_a^b f(x) dx \cong \int_a^b f_2(x) dx$$

Si se designan $a$ y $b$ como $x_0$ y $x_2$, y $f_2(x)$ se representa por un polinomio de Lagrange de segundo grado, la integral se transforma en

$$ I \cong \int_a^b \left[ \frac{(x - x_1)(x-x_2)}{(x_0 - x_1)(x_0 - x_2)}f(x_0) + \frac{(x - x_0)(x-x_2)}{(x_1 - x_0)(x_1 - x_2)}f(x_1) + \frac{(x - x_0)(x-x_1)}{(x_2 - x_0)(x_2 - x_1)}f(x_2)\right]dx$$

Después de la integración y de las manipulaciones algebraicas, se obtiene la siguiente fórmula:

$$I = \frac{h}{3}[f(x_0) + 4f(x_1) + f(x_2)] $$

Donde, en este caso, $h = (b – a)/2$. Esta ecuación se conoce como *regla de Simpson 1/3* (Chapra, S. C. y P. Canale, R. 2007. pp. 632). La regla de Simpson 1/3 también se puede expresar en terminos de $a$ y $b$

$$I = \frac{b-a}{6}[f(x_0) + 4f(x_1) + f(x_2)] $$

extendiendo este procedimiento para el caso de aplicaciones multiples, se obtiene

$$ I = \frac{b - a}{3n} \left(f(x_0) + 2 \sum_{i=1}^{n-2}f(x_{2i}) + 4 \sum_{i=1}^{n-1}f(x_{2i-1}) + f(x_n)\right)$$

Se debe utilizar un número par de segmentos
para implementar el método (Chapra, S. C. y P. Canale, R. 2007. pp. 621)

##**Regla de Simpson 3/8**

De manera similar a la obtención de la regla del trapecio y Simpson 1/3, es posible ajustar un polinomio de Lagrange de tercer grado a cuatro puntos e integrarlo:

$$ I = \int_a^b f(x) dx \cong \int_a^b f_3(x) dx$$

se obtiene

$$I \cong \frac{3h}{8}[f(x_0) + 3f(x_1) + 3f(x_2) + f(x_3)] $$

donde $h = (b – a)/3$. Esta ecuación se llama *regla de Simpson 3/8* debido a que $h$ se
multiplica por $3/8$. Ésta es la tercera fórmula de integración cerrada de Newton-Cotes. La regla $3/8$ se expresa también en la forma

$$I \cong \frac{b - a}{8}[f(x_0) + 3f(x_1) + 3f(x_2) + f(x_3)] $$

En general se tiene:

$${\displaystyle I\approx {\frac {3h}{8}}\left[f(x_{0})+3f(x_{1})+3f(x_{2})+2f(x_{3})+3f(x_{4})+3f(x_{5})+2f(x_{6})+\dots +f(x_{n})\right]}$$

(Chapra, S. C. y P. Canale, R. 2007. pp. 621)
